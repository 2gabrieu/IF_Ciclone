```python
import matplotlib.pyplot as plt #biblioteca para gerar Grafico
import ipywidgets as widgets
```

```python
k = 0.041;     # constante tabelada Stairmand
p = 0.00125;   # densidade do ar [g/cm³]
u = 0.196;     # Viscosidade Cinematica do Ar [cm²/s]
pp = 1.3;      # densidade da particula [g/cm³]
d = 20;        # Diametro do Ciclone [cm]
q = [];        # Array Vazão
f = [];        # Array Diametro de corte
v = [];        # Array Velocidade
```

```python
for i in range(5, 31, 1):                                    #Intervalo de Velocidades 5 a 30[m/s]
    q.append(i*(0.125*d**2)*100);                            #Vazão de entrada [cm³/s]
    f.append((k*d*((u*d)/(q[i-5]*((pp/p)-1))**(1/2))*10000));    #Diametro de Corte [µm]
    v.append(i);                                             #Velocidade [m/s]
```

```python
plt.plot(v,f)
plt.axis([4, 31, 0, 10])
plt.xlabel('Velocidade [m/s]')
plt.ylabel('Diametro de corte [µm]')
plt.suptitle('Stairmand')
plt.show()
```

```python
n = [];   # Array eficiencia individual de coleta
z = [];   # Array diametro de Particula
w = 10;   # Velocidade a qual sera operado o equipamento
```

```python
# Eficiencia de Coleta por diametro de particula

for j in range(1, 100, 1):                      # particulas de 1µm a 100µm
    n.append(((j/f[w-5])**2)/(1+(j/f[w-5])**2)); # Calcula a Eficiencia da particula j
    z.append(j);                                 # Tamanho da particula j
```

```python
plt.plot(z,n)

plt.xlabel('Diametro de Particula [µm]')
plt.ylabel('Eficiencia de Coleta')
plt.suptitle('Stairmand')
plt.show()
```
