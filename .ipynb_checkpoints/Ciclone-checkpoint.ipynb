```python
from ipywidgets.widgets import Label, FloatSlider, IntRangeSlider, Button 
from ipywidgets.widgets import Layout, HBox, VBox, Dropdown
from IPython.display import display
import numpy as np
import bqplot as bq
import time
import threading
```

```python
q = [1,1,1,1];        # Array Vazão

y = [5,5,5,5];        # Array Diametro de corte

x = [5,6,7,8,9,10];        # Array Velocidade

flag = True

sair = Button(
    description='Sair',
    button_style='warning', # 'success', 'info', 'warning', 'danger' or ''
)
def sair_click(b):
    global flag
    flag = False
sair.on_click(sair_click)

                                              # constante ciclone
k = Dropdown(
    options=[('Selecionar', 0), ('Stairmand', 0.041), ('Lapple', 0.095)],
    value=0
)
  

                                             # densidade do ar [kg/m³]
    
p = FloatSlider(
    value=1.25,
    min=1,
    max=1.5,
    step=0.0001,
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.4f'
)   

                                             # Viscosidade Cinematica do Ar [cm²/s]
    
u = FloatSlider(
    value=0.196,
    min=0,
    max=0.3,
    step=0.0001,
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.4f'
)   

                                             # densidade da particula [g/cm³]
    
pp = FloatSlider(    
    value=1.3,
    min=0,
    max=5,
    step=0.01,
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='.2f'
)   

                                             # Diametro do Ciclone [cm]
    
d = FloatSlider(        
    value=20,
    min=1,
    max=50,
    step=1,
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)   

ve = IntRangeSlider(
    value=[5, 20],
    min=1,
    max=30,
    step=1,
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

x_sc = bq.LinearScale(min = 0 , max = 30 )
y_sc = bq.LinearScale(min = 0 , max = 50 )

x_ax = bq.Axis(
    label = 'Velocidade',
    scale = x_sc
)

y_ax = bq.Axis(
    label = 'Diametro de corte',
    scale = y_sc,
    orientation = 'vertical'
)
line = bq.Lines(
    x = x,
    y = y,
    scales = {'x':x_sc,'y':x_sc}

)
fig = bq.Figure(marks=[line], axes=[x_ax, y_ax], title='Ciclone')
fig.max_aspect_ratio = 1
fig.min_aspect_ratio = 1

ps = HBox([Label('Densidade do Ar [kg/m³]'), p ])
us = HBox([Label('Viscosidade Cinematica [cm²/s]'), u ])
pps = HBox([Label('Densidade da Particula [g/cm³]'), pp])
ds = HBox([Label('Diametro do Ciclone [cm]'), d ])
ves = HBox([Label('Velocidade [m/s]'), ve ])
ks = HBox([Label('Modelo Ciclone'), k ])
box = VBox(children = (ks, ps, us, pps, ds, ves) )
app = VBox(children = (box, sair) )


new = VBox(children = (fig, app))

def work():
    global q
    global y
    global x
    
    while flag:
        pa = p.value
        ua = u.value
        ka = k.value
        da = d.value
        ppa = pp.value
        vea = ve.value[0]
        veb = ve.value[1]
       
        q.clear()
        y.clear()
        x.clear()
        if k.value == 0.041:
            qd = 0.125    
        else:
            qd = 0.1

        for i in range(vea, veb, 1):                        #Intervalo de Velocidades 5 a 30[m/s]
            q.append(i*(qd*da**2)*100);                            #Vazão de entrada [cm³/s]
            y.append((ka*da*((ua*da)/(q[i-vea]*((ppa/(pa*0.001))-1))**(1/2))*10000));
            x.append(i); 
            
            line.x = x
            line.y = y
        time.sleep(1)
```

```python
flag = True

thread = threading.Thread(target = work)

display(new)

thread.start()
```
